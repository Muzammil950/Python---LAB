# For Loop 

```python
l = [1,2,3,4,5,6,7,8]
```

```python
for i in l :
    print(i)
```

```python
l[0] + 1
```

```python
l1 = []
for i in l:
    print(i+1)
```

```python
l1 = []
for i in l:
    l1.append(i + 1)
l1
```

```python
l = ["asif", "arif", "azeem", "akhtar"]
```

```python
l1 = []
for i in l :
    l1.append(i.upper())
l1
```

```python
l = [23,45,90.3, 'akhtar', ' azeeem','has','that', 34,345,33243]
```

```python
l1_num = []
l2_str = []
for i in l :
    if type(i) == int or type(i) == float:
        l1_num.append(i)
    else :
        l2_str.append(i)
```
