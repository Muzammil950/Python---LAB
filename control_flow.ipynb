# if else

```python
marks = 55
if marks >= 80 :
    print("you will be a part of A0")
elif marks >= 60 and marks < 80:
    print("you will be a part of A1")
elif marks >= 40 and marks < 60:
    print("you will be a part of A2")
else :
    print("you will be a part of A3")
```

```python
marks = int(input("Enter marks : "))
if marks >= 80 :
    print("you will be a part of A0")
elif marks >= 60 and marks < 80:
    print("you will be a part of A1")
elif marks >= 40 and marks < 60:
    print("you will be a part of A2")
else :
    print("you will be a part of A3")
```
