```python
# While Loop
```

```python
a  = 1

while a <= 10:
    print(a)
    a = a + 1
```

```python
n = int(input("enter your limit"))
starting_point = 0
counter = 1

while counter <= n :
    starting_point = starting_point + counter
    counter = counter + 1

starting_point
```

```python
number = int(input("enter your number"))
factorial = 1

while number > 0 :
    factorial = factorial * number
    number = number - 1

factorial
```

```python
number = int(input("enter your number : "))
a, b = 0, 1
counter = 0

while counter < number : 
    print(a)
    c = a + b 
    a = b
    b = c 
    counter = counter + 1
```
